```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.16f}'.format
```

# REES46

```python
df_REES46_120 = pd.read_csv('.../resultados/Dados_resultados_120_janelas - REES46_120.csv')
df_REES46_120
```

```python
df_REES46_120.plot(x='Janelas', y=['IKNN', 'SKNN', 'V-SKNN', 'S-SKNN', 'SF-SKNN'], figsize=(9,5))

plt.xlabel('Janelas de avaliaÃ§Ã£o', fontsize=16)
plt.ylabel('MRR', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('REES46: MRR ao longo das janelas de avaliaÃ§Ã£o.', fontsize = 20)
plt.savefig('REES46_120_MRR.pdf', bbox_inches='tight')  
plt.show()
```

```python
ax = df_REES46_120.plot(x='Janelas', y=['IKNN (HR)', 'SKNN (HR)', 'V-SKNN (HR)', 'S-SKNN (HR)', 'SF-SKNN (HR)'], figsize=(9,5))
plt.xlabel('Janelas de avaliaÃ§Ã£o', fontsize=16)
plt.ylabel('HR', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('REES46: HR ao longo das janelas de avaliaÃ§Ã£o.', fontsize = 20)

legend_labels = ['IKNN', 'SKNN', 'V-SKNN', 'S-SKNN', 'SF-SKNN']
ax.legend(legend_labels)
plt.savefig('REES46_120_HR.pdf', bbox_inches='tight')  
plt.show()
```
